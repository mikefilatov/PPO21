## Прикладное программное обеспечение


## Семинар 2

### Задача 1

Напишите программу, которая просит пользователя ввести положительное число, и если пользователь вводит положительное число, выводит на экран сообщение "Молодец!", если нет ‒ "Это не положительное число.".

Считайте, что пользователь в любом случае вводит числа, а не какой-то текст.

### Задача 2

Напишите программу, которая принимает на вход оценку от 1 до 10, выводит на экран оценку, а после нее, с новой строки, комментарий "Плохо", "Удовлетворительно", "Хорошо", "Отлично" в зависимости от оценки.

### Задача 3

Напишите программу, которая запрашивает у пользователя пароль, и далее:

- если пароль верный, выводит на экран сообщение "Login success".
- если пароль неверный, выводит на экран сообщение "Incorrect password, try again!"

### Задача 4

Пользователь сохраняет свои любимые числа в переменные `a`, `b`, `c`

Напишите программу, которая просит пользователя ввести целое число, и если это число у пользователя любимое, на экран будет выводиться сообщение: "Мое любимое число!", если нет ‒ "Эх, ну почему?".

### Задача 5

Напишите программу, которая просит пользователя ввести число и, если оно чётное, выводит на экран сообщение: "Это число чётное", если нечётное ‒ "Это число нечётное".

### Задача 7

Влад очень любит порефлексировать над собой и окружающим миром, поэтому, чтобы пойти или не пойти на прогулку, ему нужно хорошо подумать и взвесить все за и против. Цепочка рассуждений у Влада такая:

* Сначала Влад оценивает своё настроение по 10-балльной шкале. Если его настроение 9-10, то он точно пойдет на прогулку, если от 1-4, то прогулки сегодня не будет. Иначе нужно подумать еще.

* Затем Влад смотрит на погоду. Если погода хорошая, то Влад идет гулять, если погода плохая, Влад начинает размышлять глубже

* Теперь решение о прогулке снова зависит от настроения, если настроение у Влада от 6 до 8, то он идет на прогулку, в противном случае Влад думает дальше

* Если Влад не выходил на прогулку больше 3 дней, то он пойдет на прогулку, в противном случае нужно снова подумать

* Если сейчас выходной день, то Влад идет на прогулку, иначе он остается дома, чтобы порефлексировать

Пользуясь информацией, которая дана выше, реализуйте логику размышлений Влада. Определите количество необходимых переменных для решения задачи и напишите код, который поможет Владу определиться с прогулкой.
