## Задача 1

Напишите программу, которая на вход принимает слова через пробел, а на выход выдает количество букв в этих словах

__Пример__ 

__Входные данные__

бочка стол

__Выходные данные__

9

## Задача 2

Напишите программу, которая получает на вход строку чисел с запятой в качестве разделителя. На выходе программа должна выводить среднее [гармоническое этих чисел](https://ru.wikipedia.org/wiki/%D0%A1%D1%80%D0%B5%D0%B4%D0%BD%D0%B5%D0%B5_%D0%B3%D0%B0%D1%80%D0%BC%D0%BE%D0%BD%D0%B8%D1%87%D0%B5%D1%81%D0%BA%D0%BE%D0%B5). 

__Пример__ 

__Входные данные__

10, 10, 60, 20

__Выходные данные__

15.0

## Задача 6

Магазин решил подытоживать сумму чека в разрезе категорий. Магазин торгует всего тремя категориями товаров: `овощи и фрукты`, `мясо`, `сладости`. Напишите программу, которая подытоживает сумму в чеке по категориям на основании чека по позициям. 

Программа получает на вводе строку формата `продукт, категория, цена, количество` до тех пор, пока пользователь не введет стоп-символ `_`. Затем программа выдает для каждой категории итог

__Пример__

__Входные данные__ 

курица, мясо, 230, 2

бананы, овощи и фрукты, 68, 2.4

говядина, мясо, 465, 3.3

__Выходные данные__

мясо 1994.5

овощи и фрукты 163.2
